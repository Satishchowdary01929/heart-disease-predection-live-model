```python
import numpy as np
import pandas as pd
import gradio as gr
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Load your dataset (replace with your file path)
df = pd.read_csv("drive/MyDrive/health care diabetes.csv")

# Separate features and target variable
X = df.drop("Outcome", axis=1)
y = df["Outcome"]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Train the Logistic Regression Model
model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, y_train)

# Prediction Function
def predict_diabetes(Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigree, Age):
    input_data = np.array([[Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigree, Age]])
    input_data_scaled = scaler.transform(input_data)
    prediction = model.predict(input_data_scaled)
    return "Diabetic" if prediction[0] == 1 else "Not Diabetic"

# Create the Gradio Interface with Enhancements
inputs = [
    gr.Number(label="Pregnancies"),  # Remove 'default'
    gr.Number(label="Glucose"),       # Remove 'default'
    gr.Number(label="Blood Pressure"),# Remove 'default'
    gr.Number(label="Skin Thickness"),# Remove 'default'
    gr.Number(label="Insulin"),      # Remove 'default'
    gr.Number(label="BMI"),          # Remove 'default'
    gr.Number(label="Diabetes Pedigree Function"), # Remove 'default'
    gr.Number(label="Age")           # Remove 'default'
]

# Set default values using the 'value' parameter
for i, default_value in enumerate([0, 100, 70, 20, 75, 25, 0.5, 30]):
    inputs[i].value = default_value

outputs = gr.Textbox(label="Prediction")

app = gr.Interface(
    fn=predict_diabetes,
    inputs=inputs,
    outputs=outputs,
    title="Diabetes Prediction",
    description="Enter the details to predict diabetes.",
    theme=gr.themes.Soft(),  # Use a softer theme
    examples=[
        [0, 120, 75, 25, 100, 28, 0.6, 40],
        [2, 180, 85, 30, 150, 32, 0.8, 50],
    ],
    allow_flagging="manual",
    flagging_options=["Incorrect", "Correct"],
)

# Launch the Gradio App
app.launch()
```
